```python
import pandas as pd
import numpy as np



# Replace missing values in categorical columns with mode
for col in health_df.select_dtypes(include=['object']):
    health_df[col].fillna(health_df[col].mode()[0], inplace=True)

# Custom function to calculate WOE
def calculate_woe(health_df, column, target):
    event_total = health_df[target].sum()
    non_event_total = health_df.shape[0] - event_total
    event_group = health_df.groupby(column)[target].sum()
    non_event_group = health_df.groupby(column)[target].count() - event_group
    woe = np.log((non_event_group / non_event_total) / (event_group / event_total))
    return woe

# Calculate WOE for categorical variables
#woe_values = {}
#for col in health_df.select_dtypes(include=['object']):
#    woe_values[col] = calculate_woe(health_df, col, 'Target')

#print("WOE Values:")
#print(woe_values)
```
